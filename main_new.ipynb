```python
import matplotlib.pyplot as plt

def draw_pie_chart_by_column(column):
    final_data = {}
    total = 0
    for x in column:
        if x in final_data:
            final_data[x] += 1
        else:
            final_data[x] = 1
        total += 1
    vals = final_data.values()
    anss = final_data.keys()
    wedges, _, _ = plt.pie(vals, autopct=lambda pct: f"{pct}", textprops={'color': 'w'})
    plt.legend(wedges, anss,
          title="Answers",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))
    plt.show()
```

```python
draw_pie_chart_by_column(data['Junior']['interesting_work'])
```
